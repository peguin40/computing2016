```python
# Write a program to work out how many days you have lived for.
import datetime
today=datetime.date.today()
dob=input("Enter Dob(dd/mm/yyyy): ")
dob=dob.split("/")
dob=datetime.date(int(dob[2]),int(dob[1]),int(dob[0]))
days_lived=today-dob
print("You lived for",days_lived.days,"days. ")
```

```python
# Write a program to work out how many seconds you have lived for.
import datetime
today=datetime.date.today()
dob=input("Enter Dob(dd/mm/yyyy): ")
dob=dob.split("/")
dob=datetime.date(int(dob[2]),int(dob[1]),int(dob[0]))
days_lived=today-dob
print("You lived for",days_lived.days*24*60*60,"seconds. ")
```

```python
# Write a program that will accept a date of birth and determine whether one 
# is considered an adult (i.e. 21 years old).
import datetime
today=datetime.date.today()
dob=input("Enter Dob(dd/mm/yyyy): ")
dob=dob.split("/")
dob=datetime.date(int(dob[2]),int(dob[1]),int(dob[0]))
days_lived=today-dob
years_lived=days_lived.days//365
if years_lived>=21:
    print("He/She is an adult")
else:
    print("He/She is not an adult")
```

```python
# Write a program that will generate a random playing card e.g. '9 Hearts',
# 'Queen Spades' when the return key is pressed.
# Note: Rather than generate a random number from 1 to 52. Create two random 
# numbers â one for the suit and one for the card.
suit=["Spades","Diamonds","Hearts","Clubs"]
card=["Ace","2","3","4","5","6","7","8","9","10","Jack","Queen","King"]
import random
while True:
    user_input=input()
    if user_input=="exit":
        break
    else:
        card_chosen=card[random.randint(0,12)]
        suit_chosen=suit[random.randint(0,3)]
        print(card_chosen,suit_chosen)
```

```python
# Make a game of rock, paper, scissors against the computer.
# Algorithm
# Tell user to enter either rock,paper or scissors
# Get the response
# Generate a random number from 1 to 3 (1=rock,2=paper,3=scissors)
# Compare user selection and computer selection
# Display who wins.
import random
user=input("Enter rock/paper/scissors")
choices={'rock':1,'paper':2,'scissors':3}
user=choices[user]

computer=random.randint(1,3)
if user==computer:
    print("Draw! ")
elif user==1 and computer ==3:
    print("User Won! ")
elif user>computer:
    print("User Won! ")
else:
    print("Computer Won! ")
```

```python
# Write a program that will give the answer to logic gate questions e.g.
# Enter logic gate : OR
# Enter first input : 1
# Enter second input :0
# Result = 1
# It should work for the logic gates OR, XOR, NAND and NOR gates.

# input
first = int(input("First input"))
second = int(input("Second input"))

# OR gate
if first + second >= 1:
    print('Result = 1')
else:
    print('Result = 0')

# XOR gate
if first + second == 1:
    print('Result = 1')
else:
    print('Result = 0')

# NAND Gate
if first + second = 2:
    print('Result = 1')
else:
    print('Result = 0')
    
# NOR gate
if first + second == 0:
    print('Result = 1')
else:
    print('Result = 0')
```

```python
# Write a program that will display all the factors of an input number that
# are bigger than 1.
# (e.g. the factors of the number 12 are 6,4,3 and 2 because they divide
# into 12 exactly).
# Hint: To find out whether a number X is a factor of Y use:
# If Y mod X == 0 (there is nothing remaining when Y is divided by X)

number=int(input("Input an integer: "))
factors=[]
for i in range(2,int(number**0.5+1)):
    if number%i==0:
        factors.append(i)
        factors.append(int(number/i))
for item in factors:
    print(item,end=', ')
```
